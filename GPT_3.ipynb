## text-davinci-002
The "**text-davinci-002**" model is considered as the most advanced model among the GPT-3 models that OpenAI provides, it's known to have the best generation capabilities compared to the other models, it can also complete text, with prompt provided, with high accuracy and fluency. Additionally, this model is considered to be the best choice if the task requires high language complexity and fluency.


**max_tokens**: This parameter controls the maximum number of tokens (words or word pieces) that the model should generate in its response. In this case, it's set to 2048 which means the model will generate a maximum of 2048 tokens in its response.

**n**: This parameter controls the number of responses generated by the model. In this case, it's set to 1 which means the model will generate only one response.

**stop**: This parameter allows you to specify a sequence of tokens that will stop the model from generating further. If None, the model will generate text until the max_tokens is reached.

**temperature**: This parameter controls the level of randomness or creativity of the model's response. A temperature of 0 will generate the most likely response, while a temperature of 1 will generate a more random response. In this case, it's set to 0.5 which is the default value, it means that the model will generate a response that is somewhere in between the most likely and the most creative response.

```python
prompt = "What is a docker container?"

# Set the model to "text-davinci-002" (GPT-3 model)
model = "text-davinci-002"

# Send the prompt to the API and get the response
response = openai.Completion.create(engine=model, prompt=prompt, max_tokens=2048, n=1,stop=None,temperature=0.5)

# The API returns a list of possible responses
answer = response.choices[0].text

print(response)
```

```python
def group_similar_skus(skus):
    grouped_skus = {}
    for sku in skus:
        response = openai.Completion(
            prompt=f"What other names are similar to {sku}?",
            model="text-davinci-002",
            max_tokens=2048,
            temperature=0.5
        )
        similar_skus = response["choices"][0]["text"].strip().split("\n") if response["choices"] else []
        for similar_sku in similar_skus:
            if similar_sku in grouped_skus:
                grouped_skus[similar_sku].append(sku)
            else:
                grouped_skus[similar_sku] = [sku]
    return grouped_skus

# Run the function on the 'sku' column of the DataFrame
skus = data['SKU'].tolist()
grouped_skus = group_similar_skus(skus)
```
